### Importing Modules

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Importing Iris dataset
#### For the program to run, make sure to copy both the notebook and the data into your working folder

```python
data = pd.read_csv('iris.data')
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
data.head()
```

### Summary of Iris Dataset

```python
data.describe()
```

### Summary of each variable
##### The write function will create the text file named 'iris_summary_stats.txt' where the summary of each variable will be saved

```python
sepal_len = data['sepal_length'].describe()
print(sepal_len)

#sepal_len.to_csv('iris_summary_stats.txt', sep='\t')
with open('iris_summary_stats.txt', 'w') as fl:
    fl.write(sepal_len.to_csv(sep='\t')+'\n')
```

```python
sepal_wid = data['sepal_width'].describe()
print(sepal_wid)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(sepal_wid.to_csv(sep='\t')+'\n')
```

```python
petal_len = data['petal_length'].describe()
print(petal_len)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(petal_len.to_csv(sep='\t')+'\n')
```

```python
petal_wid = data['petal_width'].describe()
print(petal_wid)
with open('iris_summary_stats.txt', 'a+') as fl:
    fl.write(petal_wid.to_csv(sep='\t')+'\n')
```

### Plotting Historgram for each variable
#### The 'Savefig' function create a png file for each variable

```python
sep_len = data['sepal_length']
plt.hist(sep_len)
plt.savefig('sep_len.png')
```

```python
sep_wid = data['sepal_width']
plt.hist(sep_wid)
plt.savefig('sep_wid.png')
```

```python
pet_len = data['petal_length']
plt.hist(pet_len)
plt.savefig('pet_len.png')
```

```python
pet_wid = data['petal_width']
plt.hist(pet_wid)
plt.savefig('pet_wid.png')
```
